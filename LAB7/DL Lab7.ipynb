```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
```

```python
# Load pretrained YOLOv8 model (nano version, fast)
model = YOLO("yolov8n.pt")
```

```python
from google.colab import files
uploaded = files.upload()  # choose your image
```

```python
# Replace 'image.jpg' with the actual uploaded filename
results = model.predict("yolo1.jpg", save=True)
results = model.predict("yolo2.jpg", save=True)
results = model.predict("yolo3.jpg", save=True)
results = model.predict("yolo4.jpg", save=True)

# Display results
result_img = results[0].plot()  # image with bounding boxes
plt.imshow(cv2.cvtColor(result_img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
